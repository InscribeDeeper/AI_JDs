```python
import pandas as pd
import os
import numpy as np
```

```python
path = "C:/Users/Administrator/Desktop/JD_results/JD_results/" #文件夹目录
files= os.listdir(path) #得到文件夹下的所有文件名称

k_begin = 4
k_end = 40

k_cend = k_end +1 # 迭代需要
T = pd.DataFrame(np.zeros(k_cend**2).reshape(k_cend,k_cend)).astype(str) # 初始化

D = list(range(k_cend))
col_name =  ['k'+'='+str(x) for x in D]
```

```python
col_name
```

```python
for k in range(k_begin, k_cend):
    f = path+"k="+str(k) # 获取文件夹
    docs = os.listdir(f) # 获取文件名
    col = 0
    for km in docs:
        doc =  open(f+'/'+km,"r", encoding = 'utf8') 
#         print(doc)
        T.iloc[k,col] = doc.read().split("\n\n\n")
        col += 1
        print(k, km, col) # 读取情况
```

```python
# 创建文件夹

def mkdir(path):
    if not os.path.exists(path):
        os.makedirs(path)   
        print("--- Create a file ---")
    else:
        print("--- Existed file ---")
		
file = path+'excel_result'
mkdir(file)    
```

```python

for k in range(4, k_cend):
    col_n = 1
    AA = pd.DataFrame(T.iloc[k,0])
    for km in range(k_cend):
        if len(T.iloc[k,km])>100:
            AA = pd.concat([AA, pd.DataFrame(T.iloc[k,km])], axis = 1)
            col_n+=1  # 读取有多少列K
    AA.columns = col_name[0:col_n] # 命名cols
    Result = AA.drop(columns=['k=0']) # 删除第一列重复了
#     Result.to_excel(path+'/'+str(k)+'-class'+'.xls') # 导出EXCEL
```
